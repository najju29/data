```
import numpy as np
import pandas as pd
import os
```

```
original_path='/content/drive/MyDrive/Celebal tech/New WER samples/Actual'
original_list=[]
for root_o, dirs_o, files_o in os.walk(original_path):
    for file_o in files_o:
        if file_o.endswith('.txt'):
            with open(os.path.join(root_o, file_o), 'r') as f:
                text = f.read()
                original_list.append(text)
```

```
df=pd.DataFrame(original_list,columns=['Text'])
```

```
from spacy.lang.hi import STOP_WORDS as STOP_WORDS_HI
import string
```

```
def text_preproessing(s): # Best techniques to txt cleaning.
  new_str=''
  for i in s:
    if i in string.punctuation+'।':
      pass
    else:
      new_str=new_str+i
  new_str_list=new_str.split(' ')
  final_str_list=[]
  for j in new_str_list:
    if j in STOP_WORDS_HI:
      pass
    else:
      final_str_list.append(j)
  return ' '.join(final_str_list)
```

```
text_list=list(df['Text'].apply(text_preproessing))
```

```
text_list
```

```
def combo_words_score(text_list,combo):
  '''
  text_list: List of all texts
  combo: number of words for which you are searching counts
  It returns a DataFrame with all possible words along with its count in Descending order.
  '''
  total_text=''
  for i in text_list:
    total_text=total_text+' '+i
  total_list=total_text.split(' ')
  unique_words=set(total_list)

  # For Single words.
  if combo==1:
    count_words={}
    for words in unique_words:
      count=total_list.count(words)
      count_words.update([(words, count)])
    count_df=pd.DataFrame({'Word':count_words.keys(),'Frequency':count_words.values()})
    count_df=count_df.iloc[count_df['Frequency'].nlargest(n=len(count_df)).index]
    return count_df

  # For Two words at a time.
  elif combo==2:
    count_words={}
    for word1 in unique_words:
      for word2 in unique_words:
        count=0
        for sentence in text_list:
          if word1 in sentence:
            if word2 in sentence:
              count=count+1
          else:
            pass
        count_words.update([(word1+','+word2, count)])
    count_df=pd.DataFrame({'Words':count_words.keys(),'Frequency':count_words.values()})
    count_df=count_df.iloc[count_df['Frequency'].nlargest(n=len(count_df)).index]
    return count_df
```

```
x=combo_words_score(text_list,2)
```
