```
import numpy as np
import pandas as pd
import os
```

```
original_path='/content/drive/MyDrive/Celebal tech/Project Word error/check sample'
original_dict = {}
for root_o, dirs_o, files_o in os.walk(original_path):
    for file_o in files_o:
        if file_o.endswith('.txt'):
            with open(os.path.join(root_o, file_o), 'r') as f_o:
                text_o = f_o.read()
                num=''
                for i in file_o:
                  if i not in 'sample_checked.txt':
                    num=num+i
                  else:
                    pass
                original_dict.update([(int(num), text_o)])
```

```
predicted_path='/content/drive/MyDrive/Celebal tech/Project Word error/havingerrorinwordsormissplace'
predicted_dict = {}
for root_p, dirs_p, files_p in os.walk(predicted_path):
    for file_p in files_p:
        if file_p.endswith('.txt'):
            with open(os.path.join(root_p, file_p), 'r') as f_p:
                text_p = f_p.read()
                num=''
                for i in file_p:
                  if i not in 'sample.txt':
                    num=num+i
                  else:
                    pass
                predicted_dict.update([(int(num),text_p)])
```

```
# coding=utf8

import numpy

from jiwer import wer


def get_word_error_rate(r, h, file_name):

    """
    Given two list of strings how many word error rate(insert, delete or substitution).
    """
    d = numpy.zeros((len(r) + 1) * (len(h) + 1), dtype=numpy.uint16)
    d = d.reshape((len(r) + 1, len(h) + 1))
    for i in range(len(r) + 1):
        for j in range(len(h) + 1):
            if i == 0:
                d[0][j] = j
            elif j == 0:
                d[i][0] = i

    for i in range(1, len(r) + 1):
        for j in range(1, len(h) + 1):
            if r[i - 1] == h[j - 1]:
                d[i][j] = d[i - 1][j - 1]
            else:
                substitution = d[i - 1][j - 1] + 1
                insertion = d[i][j - 1] + 1
                deletion = d[i - 1][j] + 1
                d[i][j] = min(substitution, insertion, deletion)
    result = float(d[len(r)][len(h)]) / len(r) * 100

    print_to_html(r, h, d,file_name)
    return result

def print_to_html(r, h, d, file_name):
    filename = "diff.html"
    x = len(r)
    y = len(h)

    html = '<html><body><head><meta charset="utf-8"></head>' \
           '<style>.g{background-color:#0080004d}.r{background-color:#ff00004d}.y{background-color:#ffa50099}</style>'

    while True:
        if x == 0 or y == 0:
            break

        if r[x - 1] == h[y - 1]:
            x = x - 1
            y = y - 1
            html = '%s ' % h[y] + html
        elif d[x][y] == d[x - 1][y - 1] + 1:    # substitution
            x = x - 1
            y = y - 1
            html = '<span class="y">%s(%s)</span> ' % (h[y], r[x]) + html
        elif d[x][y] == d[x - 1][y] + 1:        # deletion
            x = x - 1
            html = '<span class="r">%s</span> ' % r[x] + html
        elif d[x][y] == d[x][y - 1] + 1:        # insertion
            y = y - 1
            html = '<span class="g">%s</span> ' % h[y] + html
        else:
            print('\nWe got an error.')
            break

    html += '</body></html>'
    file_name_ext = "diff_html "+str(file_name)+".html"
    with open(file_name_ext, 'w',encoding="UTF-8") as f:
        f.write(html)
    f.close()
    print("Printed comparison to: {0}".format(file_name_ext))
    
```
